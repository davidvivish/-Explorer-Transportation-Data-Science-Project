**Step 2:** Let's begin by creating a chart that displays the average number of crashes per hour of the day. This will help us understand whether additional factors are contributing to crashes - i.e. rush hour, school dismissal time, night fall, etc.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convert 'CRASH DATE' and 'CRASH TIME' to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M')

# Time of Day Analysis
data['Hour of Day'] = data['CRASH TIME'].dt.hour

# Group by 'Hour of Day' and calculate the average number of crashes per hour
average_crashes_per_hour = data.groupby('Hour of Day').size() / data['Hour of Day'].nunique()

# Plot the average number of crashes
plt.figure(figsize=(12, 6))
sns.barplot(x=average_crashes_per_hour.index, y=average_crashes_per_hour.values)
plt.title('Average Number of crashes per Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Number of crashes')
plt.xticks(range(0, 24))
plt.show()


````

**Step 4**: Apply time series decomposition to review trends, seasonality, and residuals. New to time series analysis? Review the [Time Series Flashcard video series](https://youtube.com/playlist?list=PLNs9ZO9jGtUAqd0CNKySksPJJaOqYPMvk&feature=shared) here to learn about trends, components, and further analysis!

````python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Count the number of crashes per day, group by CRASH DATE
daily_crashes = data.groupby(data['CRASH DATE']).size()

# Set plot style
sns.set(style="darkgrid")

# Plot the daily crashes time series
plt.figure(figsize=(15, 6))
plt.plot(daily_crashes.index, daily_crashes.values, label='Daily crashes')
plt.title('Daily Motor Vehicle Collisions in NYC')
plt.xlabel('Date')
plt.ylabel('Number of Crashes')
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_crashes, model='additive', period=365)

# Plot the decomposed components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonality')
decomposition.resid.plot(ax=ax3)
ax3.set_title('Residuals')
plt.tight_layout()
plt.show()
````

The visualizations above provide valuable insights into the time series of daily motor vehicle collisions in New York City:

1. Time Series Plot: This shows the number of daily crashes over time. You might observe long-term trends, seasonal patterns, or significant outliers.

2. Decomposed Components:
  
    2.1 Trend: This graph shows the long-term trend in the data, which can indicate whether crashes are increasing, decreasing, or stable over time.

    2.2 Seasonality: This reveals any regular patterns that repeat over a specific period, such as yearly. It helps identify times of the year with higher or lower crash frequencies.

    2.3 Residuals: These are the irregular components that cannot be attributed to the trend or seasonality. They might include random or unpredictable fluctuations.

## Time Series Analysis

````python
# Temporal Analysis: Aggregating collision data by month
cyclist_collision_data['YearMonth'] = cyclist_collision_data['CRASH DATETIME'].dt.to_period('M')
monthly_cyclist_collisions = cyclist_collision_data.groupby('YearMonth').agg({
    'NUMBER OF CYCLIST INJURED': 'sum',
    'NUMBER OF CYCLIST KILLED': 'sum'
}).reset_index()

plt.figure(figsize=(10, 6))
plt.plot(monthly_cyclist_collisions['YearMonth'].astype(str), monthly_cyclist_collisions['NUMBER OF CYCLIST INJURED'], label='Cyclist Injuries', marker='o')
plt.plot(monthly_cyclist_collisions['YearMonth'].astype(str), monthly_cyclist_collisions['NUMBER OF CYCLIST KILLED'], label='Cyclist Fatalities', marker='o', color='red')

plt.title('Monthly Cyclist Injuries and Fatalities')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Incidents', fontsize=14)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
````

````python
cyclist_collision_data['Month'] = cyclist_collision_data['CRASH DATETIME'].dt.month

monthly_seasonal_trends = cyclist_collision_data.groupby('Month').agg({
    'NUMBER OF CYCLIST INJURED': 'mean',
    'NUMBER OF CYCLIST KILLED': 'mean'
}).reset_index()

monthly_seasonal_trends.sort_values('Month', inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(monthly_seasonal_trends['Month'], monthly_seasonal_trends['NUMBER OF CYCLIST INJURED'], label='Average Cyclist Injuries', marker='o')
plt.plot(monthly_seasonal_trends['Month'], monthly_seasonal_trends['NUMBER OF CYCLIST KILLED'], label='Average Cyclist Fatalities', marker='o', color='red')

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(monthly_seasonal_trends['Month'], [months[i-1] for i in monthly_seasonal_trends['Month']])

plt.title('Seasonal Trends in Cyclist Injuries and Fatalities')
plt.xlabel('Month')
plt.ylabel('Average Number of Incidents')
plt.legend()
plt.grid(True)
plt.show()
````

````python
cyclist_collision_data['Year'] = cyclist_collision_data['CRASH DATETIME'].dt.year

yearly_trends = cyclist_collision_data.groupby('Year').agg({
    'NUMBER OF CYCLIST INJURED': 'sum',
    'NUMBER OF CYCLIST KILLED': 'sum'
}).reset_index()

# Sorting by year to see better
yearly_trends.sort_values('Year', inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(yearly_trends['Year'], yearly_trends['NUMBER OF CYCLIST INJURED'], label='Cyclist Injuries', marker='o')
plt.plot(yearly_trends['Year'], yearly_trends['NUMBER OF CYCLIST KILLED'], label='Cyclist Fatalities', marker='o', color='red')

plt.title('Year-over-Year Trends in Cyclist Injuries and Fatalities')
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Number of Incidents', fontsize = 14)
plt.xticks(yearly_trends['Year'], rotation=45)  # Ensure the years are shown correctly
plt.legend()
plt.grid(True)
plt.show()
````

````python
bike_lanes_data['Installation Year'] = pd.to_datetime(bike_lanes_data['instdate'], errors='coerce', infer_datetime_format=True).dt.year
bike_lane_installations_by_year = bike_lanes_data.groupby('Installation Year').size().reset_index(name='New Bike Lane Installations')

collision_data['Year'] = pd.to_datetime(collision_data['CRASH DATE']).dt.year
yearly_collision_summary = collision_data.groupby('Year').agg({
    'NUMBER OF CYCLIST INJURED': 'sum',
    'NUMBER OF CYCLIST KILLED': 'sum'
}).reset_index()

combined_data = pd.merge(yearly_collision_summary, bike_lane_installations_by_year, left_on='Year', right_on='Installation Year', how='left')
combined_data['New Bike Lane Installations'] = combined_data['New Bike Lane Installations'].fillna(0)

#Use of Pearson correlation coefficient
correlation_injuries = combined_data['New Bike Lane Installations'].corr(combined_data['NUMBER OF CYCLIST INJURED'])
correlation_fatalities = combined_data['New Bike Lane Installations'].corr(combined_data['NUMBER OF CYCLIST KILLED'])

plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.scatter(combined_data['New Bike Lane Installations'], combined_data['NUMBER OF CYCLIST INJURED'])
plt.title('New Bike Lane Installations vs. Cyclist Injuries')
plt.xlabel('New Bike Lane Installations', fontsize = 14)
plt.ylabel('Number of Cyclist Injuries', fontsize = 14)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(combined_data['New Bike Lane Installations'], combined_data['NUMBER OF CYCLIST KILLED'])
plt.title('New Bike Lane Installations vs. Cyclist Fatalities')
plt.xlabel('New Bike Lane Installations', fontsize = 14)
plt.ylabel('Number of Cyclist Fatalities', fontsize = 14)
plt.grid(True)

plt.tight_layout()
plt.show()

print(f'Correlation between new bike lane installations and cyclist injuries: {correlation_injuries}')
print(f'Correlation between new bike lane installations and cyclist fatalities: {correlation_fatalities}')
````

````python
collisions_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/BigData_FinalProject/Motor_Vehicle_Collisions_Crashes.csv")
collisions_data['CRASH DATE'] = pd.to_datetime(collisions_data['CRASH DATE'], errors='coerce')

bike_lanes_data_path = "/content/drive/My Drive/Colab Notebooks/BigData_FinalProject/updated_nycbikeroutes2020.csv"
bike_lanes_data = pd.read_csv(bike_lanes_data_path)

cyclist_collisions = collisions_data[
    (collisions_data['NUMBER OF CYCLIST INJURED'] > 0) |
    (collisions_data['NUMBER OF CYCLIST KILLED'] > 0)
]

earliest_crash_date = cyclist_collisions['CRASH DATE'].min()
latest_crash_date = cyclist_collisions['CRASH DATE'].max()

collision_trends = []

for _, lane in bike_lanes_data.iterrows():
    street = lane['street']
    installation_date = pd.to_datetime(lane['instdate'], errors='coerce')

    if pd.isna(installation_date) or pd.isna(street):
        continue

    same_street_collisions = cyclist_collisions[
        cyclist_collisions['ON STREET NAME'].str.contains(street, case=False, na=False) |
        cyclist_collisions['CROSS STREET NAME'].str.contains(street, case=False, na=False)
    ]

    collisions_before = same_street_collisions[same_street_collisions['CRASH DATE'] < installation_date]
    collisions_after = same_street_collisions[same_street_collisions['CRASH DATE'] >= installation_date]

    days_before = (installation_date - earliest_crash_date).days
    days_after = (latest_crash_date - installation_date).days

    rate_before = collisions_before.shape[0] / days_before if days_before > 0 else 0
    rate_after = collisions_after.shape[0] / days_after if days_after > 0 else 0

    collision_trends.append({
        'Street': street,
        'Installation Date': installation_date,
        'Collision Rate Before': rate_before,
        'Collision Rate After': rate_after,
        'Collisions Before': collisions_before.shape[0],
        'Collisions After': collisions_after.shape[0],
        'Days Before': days_before,
        'Days After': days_after
    })

collision_trends_df = pd.DataFrame(collision_trends)
output_file_path = "/content/drive/My Drive/Colab Notebooks/BigData_FinalProject/collision_trends.csv"
collision_trends_df.to_csv(output_file_path, index=False)

output_file_path
````
